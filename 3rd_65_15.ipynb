```python
#15. Write a Python program for a Car Show room that sells cars of different brans of different price.
import pandas as pd

class Car:
    def __init__(self, brand, model, year, price, color):
        self.brand = brand
        self.model = model
        self.year = year
        self.price = price
        self.color = color
    
    def details(self):
        return {
            "Brand": self.brand,
            "Model": self.model,
            "Year": self.year,
            "Price": self.price,
            "Color": self.color
        }

class Showroom:
    def __init__(self):
        self.inventory = []
    
    def add_car(self, car):
        self.inventory.append(car)
    
    def view_cars(self):
        if not self.inventory:
            print("No cars currently available in the showroom.")
            return
        
        data = [car.details() for car in self.inventory]
        df = pd.DataFrame(data)
        print(df)
    
    def display_car_details(self, brand, model):
        found = False
        for car in self.inventory:
            if car.brand.lower() == brand.lower() and car.model.lower() == model.lower():
                print("Car Details:")
                for k, v in car.details().items():
                    print(f"{k}: {v}")
                found = True
                break
        if not found:
            print(f"No car found with brand '{brand}' and model '{model}'.")
    
    def sell_car(self, brand, model):
        try:
            for i, car in enumerate(self.inventory):
                if car.brand.lower() == brand.lower() and car.model.lower() == model.lower():
                    sold_car = self.inventory.pop(i)
                    print(f"Car sold: {sold_car.brand} {sold_car.model}")
                    return
            raise ValueError(f"Car {brand} {model} not found in inventory to sell.")
        except ValueError as e:
            print("Error:", e)
    
    def buy_car(self, brand, model, year, price, color):
        new_car = Car(brand, model, year, price, color)
        self.add_car(new_car)
        print(f"Car bought and added to inventory: {brand} {model}")

# Example usage
showroom = Showroom()

# Adding some cars initially
showroom.buy_car("Toyota", "Camry", 2020, 24000, "Red")
showroom.buy_car("Honda", "Civic", 2019, 22000, "Blue")
showroom.buy_car("Ford", "Mustang", 2021, 35000, "Black")

print("\nView all cars in showroom:")
showroom.view_cars()

print("\nDisplay details of Honda Civic:")
showroom.display_car_details("Honda", "Civic")

print("\nSell Toyota Camry:")
showroom.sell_car("Toyota", "Camry")

print("\nView all cars after selling one:")
showroom.view_cars()

print("\nTry selling a car that does not exist:")
showroom.sell_car("Tesla", "Model S")
```
